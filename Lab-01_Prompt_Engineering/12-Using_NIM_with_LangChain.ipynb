```python
import os
from google.colab import userdata
os.environ['NVIDIA_API_KEY'] = userdata.get('NVIDIA_API_KEY')
```

```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

llm = ChatNVIDIA(model="meta/llama-3.2-3b-instruct")

result = llm.invoke("Can AI think?")
print(result.content)
```

```python
%%time
result = llm.invoke("Do you think LLM understand what it generates? ")
print(result.content)
```

```python
%%time
result = llm.invoke("Which one is bigger between 9.9 and 0.11?")
print(result.content)
```

```python
%%time
result = llm.invoke("What is the minimum Ramanujan number?")
print(result.content)
```

### Call Embedding model with LangChain

```python
from langchain_nvidia_ai_endpoints import NVIDIAEmbeddings

client = NVIDIAEmbeddings(model="nvidia/llama-3.2-nv-embedqa-1b-v2")

embedding = client.embed_query("What is the capital of France?")
print(embedding)
```
